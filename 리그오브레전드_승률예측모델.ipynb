```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
M_df=pd.read_csv('/content/drive/MyDrive/Master_Ranked_Games.csv')
print(M_df.head())
M_df.info()
M_df.corr()
```

```python
win_data = M_df[M_df['blueWins'] == 1]
lose_data = M_df[M_df['blueWins'] == 0]#   

tower_count_dict_win = win_data['blueTowerKills'].value_counts().sort_index()
tower_count_dict_lose = lose_data['blueTowerKills'].value_counts().sort_index()#  

norm = plt.Normalize(tower_count_dict_lose.min(), tower_count_dict_lose.max())

win_colors = plt.cm.Reds(norm(win_data['blueTowerKills'].map(tower_count_dict_win)))
lose_colors = plt.cm.Reds(norm(lose_data['blueTowerKills'].map(tower_count_dict_lose)))# 

plt.figure(figsize=(10, 6))
plt.scatter(win_data['blueTowerKills'], win_data['blueWins'], color=win_colors, alpha=1)
plt.scatter(lose_data['blueTowerKills'], lose_data['blueWins'], color=lose_colors, alpha=0.8)
plt.xlabel('Tower Kills')
plt.grid()
plt.xticks(range(0, 13))
plt.yticks([0, 1], ['Lose', 'Win'])

sm_win = plt.cm.ScalarMappable(cmap='Blues', norm=norm)
sm_lose = plt.cm.ScalarMappable(cmap='Reds', norm=norm)

sm_win.set_array([])
sm_lose.set_array([])

plt.colorbar(sm_lose, label='Number of Tower Kills')

plt.show()
print(win_data['blueTowerKills'].value_counts())
print(lose_data['blueTowerKills'].value_counts())
#               .
```

```python
plt.boxplot([M_df['blueKills'], M_df['blueTowerKills']], labels=['Kills', 'Tower Kills'])

plt.show()
```

```python
def revove_outlier(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  df = df[(df[column] >=lower_bound)&(df[column] <= upper_bound)]
  return df

M_df = revove_outlier(M_df, 'blueKills')
M_df = revove_outlier(M_df, 'blueTowerKills')
plt.boxplot([M_df['blueKills'], M_df['blueTowerKills']], labels=['Kills', 'Tower Kills'])

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X= M_df[['blueKills','blueTowerKills']]
Y = M_df['blueWins']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=40)

model = RandomForestClassifier(n_estimators=10, max_depth=3, random_state=42)
model.fit(X_train, Y_train)

train_accuracy = accuracy_score(Y_train, model.predict(X_train))
test_accuracy = accuracy_score(Y_test, model.predict(X_test))

print(f'Train Accuracy: {train_accuracy}')
print(f'Test Accuracy: {test_accuracy}')

accuracy = model.score(X_test, Y_test)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X= M_df[['blueWardPlaced','gameDuraton']]
Y = M_df['blueWins']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=40)

model = RandomForestClassifier(n_estimators=10, max_depth=3, random_state=42)
model.fit(X_train, Y_train)

train_accuracy = accuracy_score(Y_train, model.predict(X_train))
test_accuracy = accuracy_score(Y_test, model.predict(X_test))

print(f'Train Accuracy: {train_accuracy}')
print(f'Test Accuracy: {test_accuracy}')

accuracy = model.score(X_test, Y_test)
print(f'Accuracy: {accuracy}')

confusion_matrix = confusion_matrix(Y_test, model.predict(X_test))
print('Confusion Matrix:')
print(confusion_matrix)

plt.figure(figsize=(20, 10))
plot_tree(model.estimators_[0], feature_names=X.columns, filled=True)
plt.title('First Decision Tree in the Random Forest')
plt.show()
```
